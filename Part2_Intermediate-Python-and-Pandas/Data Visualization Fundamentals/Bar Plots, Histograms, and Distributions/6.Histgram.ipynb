```python
import pandas as pd
import matplotlib.pyplot as plt

bike_sharing = pd.read_csv(r'.\Bike-Sharing-Dataset\day.csv')
bike_sharing['dteday'] = pd.to_datetime(bike_sharing['dteday'])
plt.hist(bike_sharing['casual'])
plt.title('Casual hist')
plt.show()
```

2. Try to make a few quick observations about the histogram.
   - What's the approximate range?
   0~3500
   - What's the interval with the greatest frequency?
   0~500
   - What's the interval with the lowest frequency?
   3000~3500

```python
plt.hist(bike_sharing['cnt'],bins=20)
plt.title('CNT hist')
plt.show()
```

2. Try to make a few quick observations about the histogram.
   - What's the approximate range?
   0~8000
   - What's the interval with the greatest frequency?
   4000~5000
   - What's the interval with the lowest frequency?
   0~100

3. Compare the shape of the `casual` histogram with the shape of the histogram we generated for the `cnt` column. What differences do you notice?
- `casual` histogram is postive skew, frequency decrease by count of casual users , most of casual users fall in 0~1500.
- `cnt` histogram is close to normal distribution, highist frequency fall in the center 3000~6000.
