## 3. Customizing Bar Plots
We've already saved the table above into the `weekday_averages` variable, which is a `DataFrame` (you can see the code we used in the code editor).

1. Generate a bar plot using the `weekday_averages` variable:

   - The `weekday` column provides the x-coordinates.
   - The `registered` column provides the heights.
   - Change the x-tick labels using the `plt.xticks()` function.

2. The new labels should be `['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']`.
   - The x-coordinates of the labels are `[0, 1, 2, 3, 4, 5, 6]`.
   - All labels are rotated to an angle of `30` degrees.

3. Analyze the bar plot in light of our initial observations â€” does the results support the idea that registered users rent more bikes on working days?

```python
import pandas as pd
import matplotlib.pyplot as plt

bike_sharing = pd.read_csv(r'.\Bike-Sharing-Dataset\day.csv')
bike_sharing['dteday'] = pd.to_datetime(bike_sharing['dteday'])
weekday_averages = bike_sharing.groupby('weekday').mean()[['casual', 'registered']].reset_index() 

plt.bar(weekday_averages['weekday'],weekday_averages['registered'])
plt.xticks(ticks=[0, 1, 2, 3, 4, 5, 6], labels=['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'], rotation=30)
plt.title('weekday_averages')

plt.show()
```
